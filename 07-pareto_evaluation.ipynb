# Benchmark Problems - Hypervolume Calculation
- Using pygmo package

```python
import os
import numpy as np
import pandas as pd
from pygmo import hypervolume
import matplotlib.pyplot as plt

%matplotlib inline
```

# DTLZ2 with 2 objective functions

```python
problem = 'DTLZ2-10'
path = '/home/renansantos/NetBeansProjects/PhD_MOEAD/RandomSolutions'
```

## Pareto with 10 solutions
- Literature Hypervolume = 120.178966

```python
file_name = 'pareto-10.csv'
pareto = pd.read_csv(os.path.join(path, problem, file_name), header=None)
pareto.columns = 'F1 F2'.split()

pareto_values = [list(solution) for solution in pareto.values]
hv = hypervolume(pareto_values)
ref_point = [11,11]
pareto_hv = hv.compute(ref_point)
print('Hypervolume:', pareto_hv)

pareto.plot(kind='scatter', x='F1', y='F2',figsize=(8,6))
plt.show()
```
