### Importing main libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Defining data reading and plotting methods

```python
def get_data(path,file_name):
    with open(path+file_name) as file:
        x = []
        y = []
        for line in file:
            try:
                splitted_line = line.split('Objectives:')[1].split('\t')[0].strip().split()
                x.append(float(splitted_line[0]))
                y.append(float(splitted_line[1]))
            except:
                pass
    
    df = pd.DataFrame(data=[x,y]).T
    df.columns = 'f1 f2'.split() 
    return df

def plot_pareto(df):
#     plt.figure(figsize=(8,6))
    df.plot(kind='scatter',x='f1',y='f2',alpha=0.5,figsize=(8,6))
    plt.xlabel('f1',size=15)
    plt.xticks(size=15)
    plt.ylabel('f2',size=15)
    plt.yticks(size=15)
    plt.show()
```

### Defining files path

```python
path = '/home/renansantos/NetBeansProjects/PhD_MOEAD/Results/'
```

## DTLZ1

```python
file = 'DTLZ1-2-solutions.csv'
df = get_data(path, file)
plot_pareto(df)
```

## DTLZ2

```python
file = 'DTLZ2-2-solutions.csv'
df = get_data(path, file)
plot_pareto(df)
```
