# Parallel Coordinates Plots
- this is an experiment plot in a parallel coordinates of a population of 9 objective functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import MinMaxScaler
```

```python
file1 = 'nsga_pareto_final.csv'
file2 = 'nsga_pareto_9fo_2fo.csv'

data1 = pd.read_csv(file1, sep=',',header=None)
data1.columns = 'F1 F2 F3 F4 F5 F6 F7 F8 F9'.split()
data1['Class'] = 'Pareto 1'

data2 = pd.read_csv(file2, sep=',',header=None)
data2.columns = 'F1 F2 F3 F4 F5 F6 F7 F8 F9'.split()
data2['Class'] = 'Pareto 2'
```

```python
plt.figure(figsize=(8,6))
parallel_coordinates(data1,'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
new_data = pd.concat([data1, data2])
plt.figure(figsize=(8,6))
parallel_coordinates(new_data,'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
plt.figure(figsize=(8,6))
parallel_coordinates(new_data.drop('F1',axis=1),'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
plt.figure(figsize=(8,6))
parallel_coordinates(new_data.drop('F3 F4 F5 F6 F7 F8 F9'.split(),axis=1),'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
scaler = MinMaxScaler()
d1 = pd.DataFrame(data = scaler.fit_transform(data1[data1.columns[:9]]))
d2 = pd.DataFrame(scaler.fit_transform(data2[data2.columns[:9]]))
d1['Class'] = 'Pareto 1'
d2['Class'] = 'Pareto 2'
```

```python
nd = pd.concat([d2, d1])

plt.figure(figsize=(8,6))
parallel_coordinates(nd,'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```
