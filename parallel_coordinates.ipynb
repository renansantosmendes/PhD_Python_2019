# Parallel Coordinates Plots
- this is an experiment plot in a parallel coordinates of a population of 9 objective functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
```

```python
data = pandas.read_csv('solutions.txt', sep=',',header=None)
data.columns = 'F1 F2 F3 F4 F5 F6 F7 F8 F9'.split()
data['Class'] = 'Population'

data2 = pandas.read_csv('nsga_pareto_9fo_2fo.csv', sep=',',header=None)
data2.columns = 'F1 F2 F3 F4 F5 F6 F7 F8 F9'.split()
data2['Class'] = 'Pareto'
```

```python
plt.figure(figsize=(8,6))
parallel_coordinates(data[:100],'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
new_data = pd.concat([data[:100], data2])
plt.figure(figsize=(8,6))
parallel_coordinates(new_data,'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
plt.figure(figsize=(8,6))
parallel_coordinates(new_data.drop('F1',axis=1),'Class')
plt.legend(fontsize = 'x-large')
plt.show()
```
