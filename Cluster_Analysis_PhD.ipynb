```python
print(__doc__)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
path = '/home/renansantos/NetBeansProjects/PhD_MOEAD/ClusterAnalysisRandomSolutions/'
number_of_bechmark_problem = 5
number_of_objectives = 5
file_name = 'DTLZ' + str(number_of_bechmark_problem)+\
'-'+str(number_of_objectives)+'-random_solutions.csv'
```

```python
with open(path + file_name) as file:
    new_objectives = []
    for line in file:
        splitted_line = line.replace('\n','').split(' Objectives:')
        if len(splitted_line) == 2:
            objectives = splitted_line[1].replace('\t','')\
                        .replace(' AlgorithmAttributes: {}', '').split()
            
            new_objectives.append([float(objective) for objective in objectives])

obj = pd.DataFrame(new_objectives)
```

```python
obj.plot(x=0, y=1, kind='scatter',figsize=(8,6),alpha=0.25)
plt.show()
```
