```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/home/renansantos/NetBeansProjects/PhD_MOEAD/ClusterAnalysisRandomSolutions/'
file_name = 'DTLZ5-random_solutions.csv'
```

```python
with open(path + file_name) as file:
    new_objectives = []
    for line in file:
        splitted_line = line.replace('\n','').split(' Objectives:')
        if len(splitted_line) == 2:
            objectives = splitted_line[1].replace('\t','')\
                        .replace(' AlgorithmAttributes: {}', '').split()
            
            new_objectives.append([float(objective) for objective in objectives])

obj = pd.DataFrame(new_objectives)
```

```python
obj.plot(x=0, y=1, kind='scatter',figsize=(8,6),alpha=0.25)
plt.show()
```
